```python
from google.colab import files
!pip install dash
from dash import Dash, html, dcc
import plotly.express as px
import pandas as pd


#Carregar o Arquivo
uploaded = files.upload()  # Escolha o arquivo CSV
df = pd.read_csv('ecommerce_estatistica.csv')
# 1. Histograma de Desconto
fig1 = px.histogram(df, x='Desconto', nbins=100,
                   title='Histograma de Desconto',
                   color_discrete_sequence=['red'])
fig1.update_layout(xaxis_title='Desconto', yaxis_title='Frequência')
fig1.show()
# 2. Dispersão Gênero e Quantidade de Vendas.
fig2 = px.scatter(df, x='Gênero', y='Qtd_Vendidos',
                 color='Gênero',
                 title='Relação entre Gênero e Quantidade Vendida',
                 color_discrete_sequence=['red', 'blue'])
fig2.update_layout(xaxis_title='Gênero', yaxis_title='Quantidade Vendida')
fig2.show()
# 3. Mapa Calor
fig3 = px.imshow(df[['Nota', 'Desconto', 'Preço']].corr(),
                 title='Mapa de Calor da Correlação',
                 color_continuous_scale='RdBu')
fig3.update_layout(coloraxis_colorbar_title='Correlação')
fig3.show()
# 4.Gráfico de Barra:AVALIAÇÃO POr GÊNERO.
fig4 = px.bar(df, x='Gênero', y='Nota',
              title='AVALIAÇÃO DO CLIENTE POR GÊNERO',
              color='Gênero',
              color_discrete_sequence=['red', 'blue'])
fig4.update_layout(xaxis_title='Gênero', yaxis_title='AVALIAÇÃO')
fig4.show()
# 5. Gráfico Pizza - Temporada.
mapeamento_temporada = {
    'não definido':'Indefinido',
    'outono/inverno':'Outono/Inverno',
    'outono-inverno':'Outono-Inverno',
    'primavera/verão':'Primavera-Verão',
    'primavera-verão':'Primavera-Verão',
    'primavera/verão/outono/inverno':'Todas as Estações',
    'primavera/verão outono/inverno':'Todas as Estações',
    'primavera-verão outono-inverno':'Todas as Estações',
    'primavera-verão - outono-inverno':'Todas as Estações',
}
df['Temporada'] = df['Temporada'].replace(mapeamento_temporada)
df = df[df['Temporada'] != '2021']
season_counts = df['Temporada'].value_counts()
fig5 = px.pie(values = season_counts.values, names=season_counts.index,title='% de Vendas por Temporada')
fig5.update_traces(textposition='inside', textinfo='percent+label')
fig5.show()
# 6. Gráfico de Densidade.
fig6 = px.density_contour(df, x='Nota', title='Densidade da Nota')
fig6.update_traces(contours_coloring='fill', contours_showlabels=True)
fig6.update_layout(xaxis_title='Nota', yaxis_title='Densidade')
fig6.show()
# 7.Gráfico de Regressão
fig7 = px.scatter(df, x='Nota', y='Preço',
                  title='Relação entre Nota e Preço',
                  trendline='ols')
fig7.update_traces
fig7.update_layout(xaxis_title='Nota', yaxis_title='Preço dos Produtos')
fig7.show()
#APP
app = Dash(__name__)
app.layout = html.Div([
    html.H1('Dashboard de Análise de Dados'),
    dcc.Graph(figure=fig1),
    dcc.Graph(figure=fig2),
    dcc.Graph(figure=fig3),
    dcc.Graph(figure=fig4),
    dcc.Graph(figure=fig5),
    dcc.Graph(figure=fig6),
    dcc.Graph(figure=fig7)
])
#executar APP
if __name__=='__main__':
  app.run(debug=True)

server = app.server



```
